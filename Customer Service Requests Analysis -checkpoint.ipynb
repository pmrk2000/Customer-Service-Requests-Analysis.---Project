### Function for looking unique values

```python
# Function for looking unique values 
# usually we can write like df['column'].unique or np.unique(df['column'])

def print_unique(dataset):
    for col in dataset:
        print("Unique values for column : ", col, ": Number of unique values:", len(dataset[col].unique()))
        print(dataset[col].unique())
        print("_"*120)      
print_unique(dataset)
```

### Function for looking unique values counts

```python
# Function for looking unique values counts
def print_unique(dataset):
    for col in dataset:
        print("Unique values for column : ", col, ": Number of unique values counts:", len(dataset[col].unique()))
        print(dataset[col].value_counts())
        print("_"*125)      
print_unique(dataset)
```

## $\color{red}{\text{b) Identify the variables with null values }}$

```python
# Identify the variables with null values 
# There are many approaches to identify dataset.isnull() or dataset.isnull.sum() or dataset.isna().sum()
dataset.isna().sum()
```

###  Filtering the values above 5% null values

```python
filter = (((dataset.isnull().sum()/dataset.shape[0])*100)>5)
dataset_1 = dataset.loc[: , filter.values]
list(dataset_1)
```

### Here Droping the columns with above 5% null values

```python
# columns with more than 5% nulls, these are the ones which we can drop directly
dataset_2 = dataset.drop(columns = dataset_1)
dataset_2.head()
```

```python
dataset_2.shape
```

### List the categorical columns for dataset_2


```python
categorical_feature_columns = list(set(dataset_2.columns) - set(dataset_2._get_numeric_data().columns))
len(categorical_feature_columns)
```

### List the numerical columns for dataset_2

```python
numerical_feature_columns = list(dataset_2._get_numeric_data().columns)
len(numerical_feature_columns)
```

### Here create a dataframe for categorical objects

```python
categorical_df = dataset_2.select_dtypes(exclude = np.number)
```

```python
len(categorical_df.columns)
```

### Here create a dataframe for numerical objects

```python
numerical_df = dataset_2.select_dtypes(include = np.number)
```

```python
len(numerical_df.columns)
```

### Identifying null values in categorical_df 

```python
(categorical_df.isna().sum()/categorical_df.shape[0])*100
```

### Identifying null values in numerical_df 

```python
(numerical_df.isna().sum()/numerical_df.shape[0])*100
```

### $\color{red}{\text{a)Utilize missing value treatment}}$

### Here using mode imputing method to fill null values with mode values in categorical_df

```python
for cols in categorical_df.columns:
    dataset_2[cols].fillna(dataset_2[cols].mode()[0],inplace = True)
```

### we successfully removing null values

```python
(categorical_df.isna().sum()/categorical_df.shape[0])*100
```

### Here using mean imputing method to fill null values with mean values in numerical_df

```python
for cols in numerical_df.columns:
    dataset_2[cols].fillna(dataset_2[cols].mean(),inplace = True)
```

### we successfully removing null values

```python
(numerical_df.isna().sum()/numerical_df.shape[0])*100
```

```python
dataset_2.isna().sum()
```
